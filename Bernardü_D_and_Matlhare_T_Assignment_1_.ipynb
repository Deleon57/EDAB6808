#DATA PROCESSING

*Handle missing values
Normalize or scale
Align Google Trends data with stock prices*



```python

df_weekly["Week"] = pd.to_datetime(df_weekly["Week"])
df_weekly = df_weekly.sort_values("Week")

df_weekly = df_weekly.set_index("Week")

full_index = pd.date_range(
    start=df_weekly.index.min(),
    end=df_weekly.index.max(),
    freq="D",
)

df_weekly = df_weekly.reindex(full_index).ffill()

df_weekly = df_weekly.reset_index().rename(columns={"index": "Week"})
```

```python
df_weekly = df_weekly.reset_index(drop=True)
df_daily        = df_daily.reset_index(drop=True)

df_weekly["Week"]  = pd.to_datetime(df_weekly["Week"])
df_daily["Date"]        = pd.to_datetime(df_daily["Date"])

df = df_weekly.merge(df_daily,
                            how="left",
                            left_on="Week",
                            right_on="Date",suffixes=('_weekly','_daily'))
```

```python
df['adjusted_daily_Buy_Gold']=df['Buy Gold_daily']*(df['Buy Gold_weekly']/100)

df['adjusted_daily_Gold_price_per_ounce']=df['Gold price per ounce_daily']*(df['Gold price per ounce_weekly']/100)

df['adjusted_daily_Gold_futures_price']=df['Gold futures price_daily']*(df['Gold futures price_weekly']/100)

df['adjusted_daily_Gold_market_price']=df['Gold market price_daily']*(df['Gold market price_weekly']/100)

df['adjusted_daily_Gold_price_forecast']=df['Gold price forecast_daily']*(df['Gold price forecast_weekly']/100)

df['adjusted_daily_Gold_spot_price']=df['Gold spot price_daily']*(df['Gold spot price_weekly']/100)

df['adjusted_daily_Gold_share_price']=df['Gold share price_daily']*(df['Gold share price_weekly']/100)

df['adjusted_daily_Sell_Gold']=df['Sell Gold_daily']*(df['Sell Gold_weekly']/100)

df['adjusted_daily_Gold_market']=df['Gold market_daily']*(df['Gold market_weekly']/100)

df['adjusted_daily_Gold_return']=df['Gold return_daily']*(df['Gold return_weekly']/100)

df['adjusted_daily_Gold_growth']=df['Gold growth_daily']*(df['Gold growth_weekly']/100)

df['adjusted_daily_Gold_mining']=df['Gold mining_daily']*(df['Gold mining_weekly']/100)

df['adjusted_daily_Gold_companies']=df['Gold companies_daily']*(df['Gold companies_weekly']/100)

df['adjusted_daily_Gold_Fields']=df['Gold Fields_daily']*(df['Gold Fields_weekly']/100)
```
